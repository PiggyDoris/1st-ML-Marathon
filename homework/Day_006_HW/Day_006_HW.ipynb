```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path, 並讀取 app_train
dir_data = './../data/'
f_app_train = os.path.join(dir_data, 'application_test.csv')
app_train = pd.read_csv(f_app_train)
app_train.head()
```

## 作業
將下列部分資料片段 sub_train 使用 One Hot encoding, 並觀察轉換前後的欄位數量 (使用 shape) 與欄位名稱 (使用 head) 變化

```python
sub_train = pd.DataFrame(app_train['WEEKDAY_APPR_PROCESS_START'])
print(sub_train.shape)
sub_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(handle_unknown='WEEKDAY_APPR_PROCESS_START')
sub_train_ohe = onehotencoder.fit_transform(sub_train).toarray()
pd.DataFrame(sub_train_ohe)
print(sub_train_ohe.shape)
print(sub_train_ohe)
```
